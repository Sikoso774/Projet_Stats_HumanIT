```python
# Supposons que votre DataFrame s'appelle usages_effect_m
# et que vous l'avez déjà chargé (par exemple, depuis un fichier CSV)

# Liste des colonnes à sommer
colonnes_a_sommer = [
    'S1.L1', 'S1.L2', 'S1.L3', 'S1.L4', 'S1.L5',
    'S2.L1', 'S2.L2', 'S2.L3', 'S2.L4', 'S2.L5',
    'S3.L1.1', 'S3.L1.2', 'S3.L2', 'S3.L3', 'S3.L4', 'S3.L5',
    'S4.L1.1', 'S4.L1.2', 'S4.L2', 'S4.L3', 'S4.L4', 'S4.L5',
    'S5.L1.1', 'S5.L1.2', 'S5.L2', 'S5.L3', 'S5.L4', 'S5.L5'
]
```

```python
# Somme des colonnes ligne par ligne
usages_effect_m['nvid'] = usages_effect_m[colonnes_a_sommer].sum(axis=1)
```

```python
# Afficher les premières lignes du DataFrame avec la nouvelle colonne
usages_effect_m[['Gender', 'Country_HDI.fin','Student_ID', 'nvid']].head() # Affiche les colonnes Student_ID et nvid
```

```python
#On regroupe les catégories qui nous intéressent :
colonnes_quiz = [
    'Quizz.1.score', 'Quizz.2.score', 'Quizz.3.score', 'Quizz.4.score', 'Quizz.5.score'
]

usages_effect_m['score_quiz'] = usages_effect_m[colonnes_quiz].sum(axis=1)
```

```python
#Suppression des colonnes inutiles
colonnes_a_supprimer = [
    'Intro.MOOC', 'Prez.sem.1', 'Prez.sem.2', 'Prez.sem.3', 'Prez.sem.4', 'Prez.sem.5',
    'Post.forum.0', 'view.forum.0', 'Post.forum.1', 'Post.forum.1.2', 'view.forum.1', 'view.forum.1.2',
    'Post.forum.2', 'Post.forum.2.2', 'view.forum.2', 'view.forum.2.2', 'Post.forum.3', 'view.forum.3',
    'Post.forum.4', 'Post.forum.4.2', 'view.forum.4', 'view.forum.4.2', 'Post.forum.5', 'Post.forum.5.2',
    'view.forum.5', 'view.forum.5.2', 'last.video', 'last.quizz',
    'Quizz.1.bin', 'Quizz.2.bin', 'Quizz.3.bin', 'Quizz.4.bin', 'Quizz.5.bin'
]

usages_effect_m = usages_effect_m.drop(columns=colonnes_a_supprimer, errors='ignore')
```

```python
usages_effect_m
```

```python
# Suppression des valeurs NaN dans la colonne 'Gender'
usages_effect_g = usages_effect_m.dropna(subset=['Gender'])
# Filtrage pour garder seulement "une femme" et "un homme"
usages_effect_g = usages_effect_m[usages_effect_m['Gender'].isin(['une femme', 'un homme'])]
```

```python
usages_effect_g.groupby(["Gender"])["nvid"].mean()
```

```python
usages_effect_g.groupby(["Gender"])["nvid"].std()
```

```python
# Création de la boîte à moustaches

# Supposons que usages_effect_m est déjà chargé

# Préparation des données pour le boxplot
genres = usages_effect_g['Gender'].unique() # Récupérer les genres uniques
donnees_par_genre = [usages_effect_m[usages_effect_g['Gender'] == genre]['nvid'] for genre in genres]

# Création du boxplot
plt.figure(figsize=(8, 6), dpi=100) # Taille de la figure

bp = plt.boxplot(donnees_par_genre, labels=genres, patch_artist=True, medianprops={'linewidth': 2})

plt.title('Nombre de vidéos visionnées par genre')
plt.xlabel('Genre')
plt.ylabel('Nombre de vidéos (nvid)')

# Personnalisation des boîtes
for box in bp['boxes']:
    box.set(color='#4286f4', linewidth=2)
    box.set(facecolor='#e0e0e0')

plt.show()
```

```python
usages_effect_idh=usages_effect_m.dropna(subset=["Country_HDI.fin"])
usages_effect_idh =usages_effect_m[usages_effect_m['Country_HDI.fin'].isin(['TH', 'H&M'])]
```

```python
#Création du boxplot

# Supposons que usages_effect_m est déjà chargé

# Suppression des valeurs NaN dans la colonne 'Country_HDI.fin'
usages_effect_m = usages_effect_m.dropna(subset=['Country_HDI.fin'])

# Création du boxplot avec Seaborn
plt.figure(figsize=(12, 6), dpi=100)  # Augmenter la largeur de la figure
sns.boxplot(x='Country_HDI.fin', y='nvid', data=usages_effect_m)
plt.title('Nombre de vidéos visionnées par Country_HDI.fin')
plt.xlabel('Country_HDI.fin')
plt.ylabel('Nombre de vidéos (nvid)')
plt.xticks(rotation=45, ha='right')  # Rotation des étiquettes pour lisibilité
plt.tight_layout() # Ajuster l'espacement pour éviter les chevauchements
plt.show()
```

```python

#Création du barplot avec écart-type

# Suppression des valeurs NaN dans la colonne 'Country_HDI.fin'
usages_effect_m = usages_effect_m.dropna(subset=['Country_HDI.fin'])

# Création du barplot avec Seaborn et affichage de l'écart-type
plt.figure(figsize=(12, 6), dpi=100)
sns.barplot(x='Country_HDI.fin', y='nvid', data=usages_effect_m, ci='sd')
plt.title('Moyenne du nombre de vidéos visionnées par Country_HDI.fin (avec écart-type)')
plt.xlabel('Country_HDI.fin')
plt.ylabel('Nombre de vidéos (nvid)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
